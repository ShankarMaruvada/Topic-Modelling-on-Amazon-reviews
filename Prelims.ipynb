```python
import pandas as pd
import gzip

def parse(path):
    g = gzip.open(path, 'rb')
    for l in g:
        yield eval(l)

def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient='index')
```

```python
df = getDF(data_dir+'reviews_Cell_Phones_and_Accessories_5.json.gz')
```

```python
import nltk as nt
noun_tags = ["NN","NNP","NNS","PRP","POS","WP","PRP$"]
```

```python
reviews = df.as_matrix(columns=df.columns[4:7])
```

```python
def tags(sentence):
    array_words = nt.word_tokenize(sentence)
    tags = nt.pos_tag(array_words)
    return tags
```

```python
def noun_words(review):
    nouns =[];
    all_tuples = tags(review)
    for one in all_tuples:
        for each in noun_tags:
            if(each == one[1]):
                nouns.append(one[0])
    return nouns
```

```python
total_nouns = []
for each in reviews[:10]:
    total_nouns.append(noun_words(each[0]))
```
